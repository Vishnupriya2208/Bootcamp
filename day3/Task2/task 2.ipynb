```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'income (USA)'   :    [7000,11000,8000,24000,450000],
    'income (Canada)': [7900, 80000, 81000, 8200, 825000],
    'income (Germany)': [7800, 7900, 80000, 810000, 8100]
}
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, palette="tab10", linewidth=2)
plt.title('income Over Time by Country')
plt.xlabel('income (Years)')
plt.ylabel('Year')
plt.legend(title='Country')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("lex.xlsx")
df.set_index("country", inplace=True)
df = df.transpose()  
df["Average"] = df.mean(axis=1)
plt.figure(figsize=(5, 5)) 
plt.plot(df["Average"], df.index, marker="o", linestyle="--", color="GREEN", linewidth=1, label="Average")
plt.ylabel("Year", fontsize=12)  
plt.xlabel("Average income (Years)", fontsize=12) 
plt.title(" Average income Trends Over Time", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.5)
plt.gca().invert_yaxis() 
plt.legend()
plt.tight_layout()
plt.show()
```
